```python
import numpy as np
from scipy.stats import chi2
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
df1 = 740
x1 = np.linspace(chi2.ppf(0.01, df1), chi2.ppf(0.99, df1), 100)

df2 = 735
x2 = np.linspace(chi2.ppf(0.01, df2),chi2.ppf(0.99, df2), 100)

x1 = x2 = np.linspace(1100, 1400, 1000)

y1 = chi2.sf(x1, df1)
y2 = chi2.sf(x2, df2)

fig, ax = plt.subplots(1,1,figsize=(10,6))
ax.plot(x1, y1, label = "end")
ax.plot(x2, y2, label = "start")

df3 = 735
a1 = 1361.5277; b1 = chi2.sf(a1, df1)
a2 = 1115.8679; b2 = chi2.sf(a2, df2)
# a3 = df3*1.2; b3 = chi2.sf(a3, df3)

ax.plot([a1, a2], [b1, b2], 'o')

ax.legend()
```

```python
df1 = 740
x1 = np.linspace(chi2.ppf(0.01, df1), chi2.ppf(0.99, df1), 100)

df2 = 735
x2 = np.linspace(chi2.ppf(0.01, df2),chi2.ppf(0.99, df2), 100)

x1 = x2 = np.linspace(0, 1400, 1000)

y1 = chi2.pdf(x1, df1)
y2 = chi2.pdf(x2, df2)

fig, ax = plt.subplots(1,1,figsize=(10,6))
ax.plot(x1, y1, label = "end")
ax.plot(x2, y2, label = "start")

df3 = 735

a1 = 1361.5277; b1 = chi2.pdf(a1, df1)
a2 = 1115.8679; b2 = chi2.pdf(a2, df2)
a3 = df3*1.2; b3 = chi2.pdf(a3, df3)
ax.plot([a1, a2, a3], [b1, b2, b3], 'o')

ax.legend()
```
